```python
number_in_letters = ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
numbers = list(map(str, range(10)))

def find_first_number(my_string):
    first_number = ""
    for char in my_string:
        if char in numbers:
            first_number = char
            return(first_number)
        first_number = first_number + char
        matches_number = [word for word in number_in_letters if word.startswith(first_number)]
        if not matches_number:
            for i in range(len(first_number)):
                first_number = first_number[1:]
                matches_number = [word for word in number_in_letters if word.startswith(first_number)]
                if first_number == "":
                    break
                elif matches_number:
                    break
            continue
        if first_number in number_in_letters:
            return(numbers[number_in_letters.index(first_number)])
```

```python
def find_last_number(my_string):
    last_number = ""
    for char in reversed(my_string):
        if char in numbers:
            last_number = char
            return(last_number)
        last_number = char + last_number
        matches_number = [word for word in number_in_letters if word.endswith(last_number)]
        if not matches_number:
            for i in range(len(last_number)):
                last_number = last_number[:-1]
                matches_number = [word for word in number_in_letters if word.endswith(last_number)]
                if last_number == "":
                    break
                elif matches_number:
                    break
            continue
        if last_number in number_in_letters:
            return(numbers[number_in_letters.index(last_number)])
```
